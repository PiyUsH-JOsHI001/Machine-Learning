# Ordinal Encoding 

```python
import pandas as pd
```

```python
data=pd.DataFrame({"size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
```

```python
data
```

```python
ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
od=OrdinalEncoder(categories=ord_data)
od.fit_transform(data[["size"]])
```

```python
data["en_size"]=od.fit_transform(data[["size"]])
```

```python
data
```

```python
# by map function
```

```python
ord_df={"s":0,"m":1,"l":2,"xl":3}
```

```python
data["size"].map(ord_df)
```

```python
data["en_size"]=data["size"].map(ord_df)
```

```python
data
```
