# Remove Outliers using IQR (Interquartile Range)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\MS OFFICE FILE\MS EXCEL FILE\heart2.csv")
df.head(10)
```

```python
df.describe()
```

```python
# plt.figure(figsize=(10,5))
sns.boxplot(x="trestbps",data=df)
plt.show()
```

```python
df.shape
```

```python
Q1=df["trestbps"].quantile(0.25)
Q1
```

```python
Q3=df["trestbps"].quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
```

```python
min_range = Q1-(1.5*IQR)
max_range = Q3+(1.5*IQR)
```

```python
min_range,max_range
```

```python
df[df["trestbps"]<=max_range]
```

```python
new_df = df[df["trestbps"]<=max_range]
```

```python
new_df.shape
```

```python
sns.boxplot(x="trestbps",data=new_df)
plt.show()
```

```python
new_df
```

```python
new_df.shape
```

# Remove Outliers using Z Score

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv(r"D:\MS OFFICE FILE\MS EXCEL FILE\heart2.csv")
ds.head(10)
```

```python
ds.isnull().sum()
```

```python
ds.describe()
```

```python
sns.boxplot(x="trestbps",data=ds)
plt.show()
```

```python
sns.distplot(ds["trestbps"])
plt.show()
```

```python
minRange = ds["trestbps"].mean()-(3*ds["trestbps"].std())
maxRange = ds["trestbps"].mean()+(3*ds["trestbps"].std())
```

```python
minRange,maxRange
```

```python
new_ds=ds[ds["trestbps"]<=maxRange]
```

```python
sns.boxplot(x="trestbps",data=new_ds)
plt.show()
```

```python
new_ds.shape
```

```python
new_ds
```
